```python
import cv2
import matplotlib.pyplot as plt
from cvlib.object_detection import draw_bbox
import cvlib
```

```python
#CARS
```

```python
img=cv2.imread("cars.jpg")
```

```python
plt.imshow(img)
```

```python
box,label,count=cvlib.detect_common_objects(img)
#detect_common_objects ile bilinen nesneleri ekiyoruz
```

```python
output=draw_bbox(img,box,label,count) #draw_box tand nesnelerin etrafna kare iziyor
#box kare
#label tand nesne
```

```python
plt.figure(figsize = (10,10))
plt.imshow(output)
```
